```python
stolen_base_data = batters_2024_raw_data[
    ['last_name, first_name', 'r_stolen_base_2b', 'r_stolen_base_3b',
     'r_stolen_base_home', 'sprint_speed', 
     'r_total_stolen_base']
]
top10fastesttosteal2b = (
    stolen_base_data
    .sort_values(by='r_stolen_base_2b', ascending=False)
    .head(10)
)
AverageSpeedNeededtoSteal2b = top10fastesttosteal2b['sprint_speed'].mean()


top10fastesttosteal3b = (
    stolen_base_data
    .sort_values(by='r_stolen_base_3b', ascending=False)
    .head(10)
)
AverageSpeedNeededtoSteal3b = top10fastesttosteal3b['sprint_speed'].mean()



top10fastesttostealhome = (
    stolen_base_data
    .sort_values(by='r_stolen_base_home', ascending=False)
    .head(10)
)
AverageSpeedNeededtoStealHome = top10fastesttostealhome['sprint_speed'].mean()

#print("Top 10 Fastest to Steal 2B:\n", top10fastesttosteal2b)
print("Average Speed needed to confidently steal second base:",AverageSpeedNeededtoSteal2b,"mph.")
print("Average Speed needed to confidently steal third base:",AverageSpeedNeededtoSteal3b,"mph.")
print("Average Speed needed to confidently steal home:",AverageSpeedNeededtoStealHome,"mph.")




```

```python
# Top 10 fastest players by sprint speed
top10sprintspeed = (
    stolen_base_data.sort_values(by='sprint_speed', ascending=False).head(10)
)

# Print only player name and sprint speed
print("\nTop 10 fastest players by sprint speed:\n")
print(top10sprintspeed[['last_name, first_name', 'sprint_speed']])

```

```python
#Correlate pitch types (e.g., fastball, slider, curveball) with opponents' batting averages (p_opp_batting_avg) to identify which pitches are most effective.
#Calculate the whiff percentage (whiff_percent) by pitch type and its relationship to spin rates or velocity.
```

```python
pitchers_2024_raw_data = pd.read_csv('~/downloads/stats (1).csv')
pitch_types = ['pitch_count_fastball', 'pitch_count_offspeed', 'pitch_count_breaking']
target_column = 'whiff_percent'  
pitchers_2024_raw_data['total_pitches'] = pitchers_2024_raw_data[
    ['pitch_count_fastball', 'pitch_count_offspeed', 'pitch_count_breaking']
].sum(axis=1)
for pitch in pitch_types:
    pitchers_2024_raw_data[f'{pitch}_percent'] = (
        pitchers_2024_raw_data[pitch] / pitchers_2024_raw_data['total_pitches']
    )
    
correlations = pitchers_2024_raw_data[
    [f'{pitch}_percent' for pitch in pitch_types] + [target_column]
].corr()[target_column][:-1]  # Correlate pitch percentages with p_opp_batting_avg

# Display results
print("Correlation with p_opp_batting_avg:")
print(correlations)

# Visualize (optional)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.barplot(x=correlations.index, y=correlations.values)
plt.title("Correlation of Pitch Types with Whiff Percent")
plt.ylabel("Correlation Coefficient")
plt.xticks(rotation=45)
plt.show()
```
